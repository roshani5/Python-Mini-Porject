# Smart user utility system
# A mini project using python basics

```python
print("===== SMART USER UTILITY SYSTEM =====")
print("1. User Info & Salary Calculation")
print("2. Eligibility Check")
print("3. Permission System")
print("4. Exit")

choice = int(input("Enter your choice (1-4): "))

if choice == 1:
    name = input("Enter your name: ")
    age = int(input("Enter your age: "))
    salary = float(input("Enter monthly salary: "))

    yearly_salary = salary * 12
    bonus = yearly_salary * 0.10
    total_income = yearly_salary + bonus

    print("\n--- USER DETAILS ---")
    print("Name:", name)
    print("Age:", age)
    print("Yearly Salary:", yearly_salary)
    print("Bonus:", bonus)
    print("Total Income:", total_income)

elif choice == 2:
    age = int(input("Enter your age: "))
    income = float(input("Enter your total annual income: "))

    if age >= 18 and income > 500000:
        print("Eligible for Premium Membership")
    else:
        print("Not Eligible for Premium Membership")

elif choice == 3:
    basic = 4
    edit = 2
    admin = 1

    print("Permission Options:")
    print("1. Basic")
    print("2. Edit")
    print("3. Admin")

    user_choice = int(input("Choose permission (1-3): "))

    if user_choice == 1:
        permission = basic
    elif user_choice == 2:
        permission = basic | edit
    elif user_choice == 3:
        permission = basic | edit | admin
    else:
        permission = 0

    print("Permission Value:", permission)

    if permission & admin:
        print("Admin Access Granted")
    else:
        print("Admin Access Not Granted")

elif choice == 4:
    print("Exiting program... Thank you!")

else:
    print("Invalid choice")
```
